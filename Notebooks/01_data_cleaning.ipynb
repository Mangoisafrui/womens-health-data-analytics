```python
# ============================================
# STEP 1: LOAD ALL DATASETS
# ============================================

base = "C:/Users/Dell/OneDrive/Desktop/Women's Mental Health/Raw Data"

# GBD Prevalence
gbd_prev = pd.read_csv(f"{base}/GBD_2021/GBD_2021_Prevalence_Women_1549.csv")

# GBD DALYs
gbd_daly = pd.read_csv(f"{base}/GBD_2021/GBD_2021_DALYs_Women_1549.csv")

# World Bank Country Classifications
world_bank = pd.read_excel(f"{base}/World_Bank/World_Bank_Country_Classifications.xlsx")

# UNDP Gender Inequality Index — Excel format despite .csv extension
gii = pd.read_excel(f"{base}/GII_UNDP/UNDP_GII_All_Countries.csv")

# WHO Psychiatrists per 100k
psychiatrists = pd.read_csv(f"{base}/WHO_Atlas/WHO_Psychiatrists_Per_100k.csv")

# World Bank GDP per Capita
gdp = pd.read_csv(f"{base}/World_Bank/World_Bank_GDP_Per_Capita.csv", skiprows=4)

# ── Confirm all loaded ──
print("✓ GBD Prevalence:", gbd_prev.shape)
print("✓ GBD DALYs:", gbd_daly.shape)
print("✓ World Bank Classifications:", world_bank.shape)
print("✓ UNDP GII:", gii.shape)
print("✓ WHO Psychiatrists:", psychiatrists.shape)
print("✓ GDP per Capita:", gdp.shape)
print("\nAll datasets loaded successfully!")
```

```python
# ============================================
# STEP 2: EXPLORE EACH DATASET
# ============================================

datasets = {
    "GBD Prevalence": gbd_prev,
    "GBD DALYs": gbd_daly,
    "World Bank": world_bank,
    "UNDP GII": gii,
    "WHO Psychiatrists": psychiatrists,
    "GDP per Capita": gdp
}

for name, df in datasets.items():
    print(f"\n{'='*50}")
    print(f"DATASET: {name}")
    print(f"Shape: {df.shape}")
    print(f"Columns: {list(df.columns)}")
    print(f"Missing values:\n{df.isnull().sum()}")
    print(f"\nFirst row:\n{df.iloc[0]}")
```

```python
# ============================================
# STEP 3: LOOK INSIDE EACH DATASET
# ============================================

# GBD Prevalence — first 3 rows
print("=== GBD PREVALENCE ===")
print(gbd_prev.head(3))

print("\n=== WORLD BANK CLASSIFICATIONS ===")
print(world_bank.head(3))
print("Columns:", list(world_bank.columns))

print("\n=== UNDP GII ===")
print(gii.head(3))
print("Columns:", list(gii.columns))

print("\n=== WHO PSYCHIATRISTS ===")
print(psychiatrists.head(3))

print("\n=== GDP PER CAPITA ===")
print(gdp.head(3))
print("First 10 columns:", list(gdp.columns[:10]))
```

```python
# ============================================
# STEP 4: CHECK COUNTRY NAME FORMATS
# ============================================

print("GBD unique countries:", gbd_prev['location'].nunique())
print("Sample GBD countries:", list(gbd_prev['location'].unique()[:5]))

print("\nWorld Bank unique countries:", world_bank['Economy'].nunique())
print("Sample World Bank countries:", list(world_bank['Economy'].unique()[:5]))

print("\nGDP unique countries:", gdp['Country Name'].nunique())
print("Sample GDP countries:", list(gdp['Country Name'].unique()[:5]))

print("\nWHO Psychiatrists columns:", list(psychiatrists.columns))
print("Sample WHO countries:", list(psychiatrists.iloc[:5, 0]))
```

```python
# ============================================
# STEP 5: CLEAN COLUMN NAMES
# ============================================

# Rename country columns to standard name
gbd_prev = gbd_prev.rename(columns={'location': 'country'})
gbd_daly = gbd_daly.rename(columns={'location': 'country'})
world_bank = world_bank.rename(columns={'Economy': 'country'})
gdp = gdp.rename(columns={'Country Name': 'country'})
psychiatrists = psychiatrists.rename(columns={
    'Entity': 'country',
    'Code': 'country_code',
    'Year': 'year',
    'Total number of psychiatrists per 100,000 population': 'psychiatrists_per_100k'
})

# Confirm
print("✓ GBD Prevalence columns:", list(gbd_prev.columns))
print("✓ World Bank columns:", list(world_bank.columns))
print("✓ GDP columns:", list(gdp.columns[:5]))
print("✓ WHO columns:", list(psychiatrists.columns))
print("\nColumn names standardized!")
```

```python
# ============================================
# STEP 6: CLEAN GBD DATASETS
# ============================================

# Keep only columns we need
gbd_prev_clean = gbd_prev[['country', 'cause', 'year', 'val']].copy()
gbd_daly_clean = gbd_daly[['country', 'cause', 'year', 'val']].copy()

# Rename val column to be descriptive
gbd_prev_clean = gbd_prev_clean.rename(columns={'val': 'prevalence_rate'})
gbd_daly_clean = gbd_daly_clean.rename(columns={'val': 'daly_rate'})

# Separate depression and anxiety
prev_depression = gbd_prev_clean[gbd_prev_clean['cause'] == 'Depressive disorders'].copy()
prev_anxiety = gbd_prev_clean[gbd_prev_clean['cause'] == 'Anxiety disorders'].copy()

daly_depression = gbd_daly_clean[gbd_daly_clean['cause'] == 'Depressive disorders'].copy()
daly_anxiety = gbd_daly_clean[gbd_daly_clean['cause'] == 'Anxiety disorders'].copy()

# Confirm
print("Depression Prevalence:", prev_depression.shape)
print("Anxiety Prevalence:", prev_anxiety.shape)
print("Depression DALYs:", daly_depression.shape)
print("Anxiety DALYs:", daly_anxiety.shape)
print("\nSample depression data:")
print(prev_depression.head(3))
```

```python
# ============================================
# STEP 7: CLEAN GDP DATASET
# ============================================

# Drop unnecessary columns
gdp_clean = gdp[['country', 'Country Code', '2000', '2005', '2010', 
                  '2015', '2019', '2020', '2021']].copy()

# Remove regional aggregates — keep only real countries
# Real countries have a 3-letter code
gdp_clean = gdp_clean[gdp_clean['Country Code'].str.len() == 3]

# Rename country code column
gdp_clean = gdp_clean.rename(columns={'Country Code': 'country_code'})

# Check result
print("GDP clean shape:", gdp_clean.shape)
print("Missing values:\n", gdp_clean.isnull().sum())
print("\nSample:")
print(gdp_clean.head(3))
```

```python
# ============================================
# STEP 8: CLEAN GDP FURTHER
# ============================================

# Remove regional aggregates using World Bank country list
# Keep only countries that exist in our World Bank classifications
valid_countries = world_bank['country'].tolist()
gdp_clean = gdp_clean[gdp_clean['country'].isin(valid_countries)]

print("GDP after removing aggregates:", gdp_clean.shape)
print("Missing values:\n", gdp_clean.isnull().sum())
print("\nSample:")
print(gdp_clean.head(3))
```

```python
# ============================================
# STEP 9: REMOVE REMAINING AGGREGATES
# ============================================

# Check what's still in there that shouldn't be
# Regional aggregates usually contain words like these
remove_keywords = [
    'Africa', 'Asia', 'Europe', 'America', 'Caribbean',
    'Pacific', 'OECD', 'World', 'income', 'dividend',
    'Middle East', 'Central', 'Eastern', 'Western',
    'Northern', 'Southern', 'Latin', 'Sub-Saharan',
    'fragile', 'blend', 'IDA', 'IBRD', 'Euro'
]

# Build a filter — remove rows where country name contains any keyword
pattern = '|'.join(remove_keywords)
gdp_clean = gdp_clean[~gdp_clean['country'].str.contains(pattern, case=False, na=False)]

print("GDP after removing all aggregates:", gdp_clean.shape)
print("\nSample countries:")
print(gdp_clean['country'].head(10).tolist())

# Fill missing GDP values with median
gdp_clean = gdp_clean.fillna(gdp_clean.median(numeric_only=True))
print("\nMissing values after fill:")
print(gdp_clean.isnull().sum())
```

```python
# ============================================
# STEP 10: CLEAN WHO PSYCHIATRISTS DATASET
# ============================================

# Check what years are available
print("Years available:", sorted(psychiatrists['year'].unique()))
print("Countries:", psychiatrists['country'].nunique())
print("\nSample:")
print(psychiatrists.head(5))
```

```python
# ============================================
# STEP 11: CLEAN WHO PSYCHIATRISTS DATASET
# ============================================

# Keep only 2019 data — most complete year
psychiatrists_clean = psychiatrists[psychiatrists['year'] == 2019].copy()

# Drop year and country_code columns — not needed for merge
psychiatrists_clean = psychiatrists_clean[['country', 'psychiatrists_per_100k']]

# Check result
print("Psychiatrists clean shape:", psychiatrists_clean.shape)
print("Missing values:\n", psychiatrists_clean.isnull().sum())
print("\nSample:")
print(psychiatrists_clean.head(5))

# Check range of values
print("\nPsychiatrists per 100k statistics:")
print(psychiatrists_clean['psychiatrists_per_100k'].describe())
```

```python
# ============================================
# STEP 12: EXPLORE GII DATASET
# ============================================

print("GII columns:", list(gii.columns))
print("\nGII shape:", gii.shape)
print("\nSample:")
print(gii.head(3))
print("\nUnique indicators:")
print(gii['indicator'].unique() if 'indicator' in gii.columns else "No indicator column")
```

```python
# ============================================
# STEP 12: EXPLORE GII DATASET
# ============================================

print("GII columns:", list(gii.columns))
print("\nGII shape:", gii.shape)
print("\nSample:")
print(gii.head(3))
```

```python
# ============================================
# STEP 13: CLEAN GII DATASET
# ============================================

# Keep only GII value rows
gii_clean = gii[gii['indicatorCode'] == 'gii'].copy()

# Keep only columns we need
gii_clean = gii_clean[['country', 'year', 'value']]

# Rename value column
gii_clean = gii_clean.rename(columns={'value': 'gii_score'})

# Keep only years 2000-2021
gii_clean = gii_clean[(gii_clean['year'] >= 2000) & (gii_clean['year'] <= 2021)]

print("GII clean shape:", gii_clean.shape)
print("Years covered:", sorted(gii_clean['year'].unique()))
print("Missing values:\n", gii_clean.isnull().sum())
print("\nSample:")
print(gii_clean.head(5))
```

```python
# ============================================
# STEP 14: RESHAPE GDP TO LONG FORMAT
# ============================================

gdp_long = gdp_clean.melt(
    id_vars=['country', 'country_code'],
    var_name='year',
    value_name='gdp_per_capita'
)

gdp_long['year'] = gdp_long['year'].astype(int)

print("GDP long shape:", gdp_long.shape)
print("\nSample:")
print(gdp_long.head(6))
```

```python
# ============================================
# STEP 15: CREATE MASTER DATASET
# ============================================

# Start with depression prevalence as base
# Start with the depression dataset as our base table — everything else will be added to this
master = prev_depression.copy()

# Merge anxiety prevalence
# Add the anxiety rate column by matching rows where country AND year are the same
master = master.merge(
    prev_anxiety[['country', 'year', 'prevalence_rate']].rename(
        columns={'prevalence_rate': 'anxiety_rate'}),
    on=['country', 'year'], how='left'
)

# Rename depression rate
master = master.rename(columns={'prevalence_rate': 'depression_rate'})

# Drop cause column
# Delete the cause column because it's now redundant — every row is already depression data
master = master.drop(columns=['cause'])

print("After merging anxiety:", master.shape)
print(master.head(3))
```

```python
# ============================================
# STEP 16: MERGE REMAINING DATASETS
# ============================================

# Merge World Bank classifications
master = master.merge(
    world_bank[['country', 'Region', 'Income group']],
    on='country', how='left'
)

# Merge GII
master = master.merge(
    gii_clean,
    on=['country', 'year'], how='left'
)

# Merge GDP
master = master.merge(
    gdp_long[['country', 'year', 'gdp_per_capita']],
    on=['country', 'year'], how='left'
)

# Merge Psychiatrists
master = master.merge(
    psychiatrists_clean,
    on='country', how='left'
)

print("Master dataset shape:", master.shape)
print("Columns:", list(master.columns))
print("Missing values:\n", master.isnull().sum())
print("\nSample:")
print(master.head(3))
```

```python
# ============================================
# STEP 17: HANDLE MISSING VALUES
# ============================================

# Option A — Keep only key years
key_years = [2000, 2005, 2010, 2015, 2019, 2020, 2021]
master_key = master[master['year'].isin(key_years)].copy()

print("Master key years shape:", master_key.shape)
print("Missing values:\n", master_key.isnull().sum())
```

```python
#  ===========================================
# STEP 18: HANDLE MISSING VALUES
# ============================================

# Block 1
master_clean = master_key.dropna(subset=['Income group']).copy()

# Block 2
master_clean['gii_score'] = master_clean.groupby('country')['gii_score'].transform(
    lambda x: x.fillna(x.median())
)

# Block 3
for col in ['gdp_per_capita', 'psychiatrists_per_100k']:
    master_clean[col] = master_clean.groupby('Income group')[col].transform(
        lambda x: x.fillna(x.median())
    )

print("✓ Missing values handled:")
print("  - Rows with missing Income group: dropped")
print("  - GII score: filled with country median")
print("  - GDP per capita: filled with income group median")
print("  - Psychiatrists per 100k: filled with income group median")
print("\nMissing values remaining:")
print(master_clean.isnull().sum())
print("\nFinal master dataset shape:", master_clean.shape)
```

```python
# ============================================
# STEP 19: DROP REMAINING MISSING GII ROWS
# ============================================

master_clean = master_clean.dropna(subset=['gii_score'])

print("Missing values after dropping:")
print(master_clean.isnull().sum())
print("\nFinal master dataset shape:", master_clean.shape)
```

```python
# ============================================
# STEP 19b: FINAL CLEAN & RESAVE
# ============================================

# Drop remaining GII nulls
master_clean = master_clean.dropna(subset=['gii_score'])

print("Missing values after final drop:")
print(master_clean.isnull().sum())
print("\nFinal shape:", master_clean.shape)

# Resave final master dataset
output_path = "C:/Users/Dell/OneDrive/Desktop/Women's Mental Health/Clean Data/master_dataset.csv"
master_clean.to_csv(output_path, index=False)

print("\n✓ Final master dataset saved!")
print("\nPreview:")
print(master_clean.head(3))
```

```python
# ============================================
# STEP 20: SAVE MASTER DATASET
# ============================================

output_path = "C:/Users/Dell/OneDrive/Desktop/Women's Mental Health/Clean Data/master_dataset.csv"

master_clean.to_csv(output_path, index=False)

print("✓ Master dataset saved successfully!")
print(f"Location: {output_path}")
print(f"Shape: {master_clean.shape}")
print(f"\nPreview:")
print(master_clean.head(5))
```

```python
# ============================================
# STEP 21: ADD DALYS TO MASTER DATASET
# ============================================

# Clean depression DALYs
daly_depression_clean = daly_depression[['country', 'year', 'daly_rate']].copy()
daly_depression_clean = daly_depression_clean.rename(columns={'daly_rate': 'depression_daly'})

# Clean anxiety DALYs
daly_anxiety_clean = daly_anxiety[['country', 'year', 'daly_rate']].copy()
daly_anxiety_clean = daly_anxiety_clean.rename(columns={'daly_rate': 'anxiety_daly'})

# Keep only key years
daly_depression_clean = daly_depression_clean[daly_depression_clean['year'].isin(key_years)]
daly_anxiety_clean = daly_anxiety_clean[daly_anxiety_clean['year'].isin(key_years)]

print("✓ Depression DALYs shape:", daly_depression_clean.shape)
print("✓ Anxiety DALYs shape:", daly_anxiety_clean.shape)
print("\nSample:")
print(daly_depression_clean.head(3))
```

```python
# ============================================
# STEP 21: MERGE DALYS INTO MASTER
# ============================================

# Merge depression DALYs
master_clean = master_clean.merge(
    daly_depression_clean,
    on=['country', 'year'],
    how='left'
)

# Merge anxiety DALYs
master_clean = master_clean.merge(
    daly_anxiety_clean,
    on=['country', 'year'],
    how='left'
)

print("Master shape after adding DALYs:", master_clean.shape)
print("Columns:", list(master_clean.columns))
print("Missing values:\n", master_clean.isnull().sum())
```

```python
# ============================================
# STEP 23: RESAVE FINAL MASTER DATASET
# ============================================

output_path = "C:/Users/Dell/OneDrive/Desktop/Women's Mental Health/Clean Data/master_dataset.csv"
master_clean.to_csv(output_path, index=False)

print("✓ Final master dataset saved!")
print(f"Shape: {master_clean.shape}")
print(f"Columns: {list(master_clean.columns)}")
```
